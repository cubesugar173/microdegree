```python
day1_price = stock.get_market_ohlcv("20230901")
day1_price.rename(columns={'종가': 'day1'}, inplace=True)
day1_price.head()
```

```python
day27_price = stock.get_market_ohlcv("20230927")
day27_price.rename(columns={'종가': 'day27'}, inplace=True)
day27_price.head()
```

```python
t_df = pd.concat([day1_price['day1'], day27_price['day27']], axis=1)
t_df.head()
```

```python
#변화량을 어케 보지.... 종가/시가*100??
t_df['ratio'] = t_df['day27'] / t_df['day1'] * 100
t_df.head()
```

```python
t_df['diff'] = t_df['ratio'] - 100
t_df.head()
```

```python
t_df['abs_diff'] = t_df['diff'].abs()
t_df.head()
```

```python
t_df = t_df.sort_values(by = 'abs_diff', ascending=False)
t_df.head()
```
